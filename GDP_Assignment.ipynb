## 5.	Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e., the states lying between the 85th and the 100th percentile are in C1; those between the 50th and the 85th percentiles are in C2, and so on

### Ans: <font color="green"> Please see the code below</font>

```python


quart=[0]
quart.append(np.percentile(df['Per Capita GSDP (Rs.)'],20))
quart.append(np.percentile(df['Per Capita GSDP (Rs.)'],50))
quart.append(np.percentile(df['Per Capita GSDP (Rs.)'],85))
quart.append(np.percentile(df['Per Capita GSDP (Rs.)'],100))
print (quart)
```

```python
def category (quart,x):
    if x >=quart[3] :
        return 'C1'
    elif x >=quart[2] :
        return 'C2'
    elif x >=quart[1] :
        return 'C3'
    else:
        return 'C4'
        
df['Category'] = df.apply(lambda x: category(quart,x['Per Capita GSDP (Rs.)']),axis=1)
df.loc[:,['State_Name','Per Capita GSDP (Rs.)','Category']]
```

```python
# Create function to tanspose each catogery datafranm and then add a column having Sum of each sub-sactor
def Convert(x):
    x=x.T
    x=x.reset_index()
    x.rename(columns=x.iloc[0]).iloc[1:,0:]
    State_Name=[]
    State_Name=x.iloc[32,1:]
    #print(State_Name)
    x=x.rename(columns=State_Name)
    x=x.loc[x['index'].isin(SC)]
    x=x.rename(columns=State_Name)
    x.set_index('index',inplace=True)
    x['Total']=x.sum(axis=1,skipna=True)
    x=x.sort_values(by='Total',ascending = False)
    #x=x['Total']
    #x['Run_Total']=x['Total'].cumsum()
    Total_GSDP_for_Catg=x.at['Gross State Domestic Product','Total']
    x['% Contribution']=x['Total']/Total_GSDP_for_Catg*100
    #x['Run_total_per']=x['Run_Total']/Total_GSDP_for_Catg*100
    x=x.drop('Gross State Domestic Product')
    return(x,Total_GSDP_for_Catg)

# Create function to add the cumilative sum and the cumilative % of each sub catogery and find Sub-Category contributing 
#approx 80% of the GSDP 
def cum(x,y):
    x['Run_Total']=x['Total'].cumsum()
    x['Run_total_per']=x['Run_Total']/y*100
    x=x[x['Run_total_per']<=80]
    return(x)

# Create afunction to plot the charts
def plot_chart(dt_frm):
    plt_labels=dt_frm.index
    size=dt_frm['Run_total_per']
    plt.pie(size, labels=plt_labels, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    #plt.title(dt_frm.describe())
    plt.show()
#Convert(Cat1_states)
#Cat1_states['Total']
#Cat1_states.sort_values(by='Run_Total',ascending = False)
```

## 2	Plotting graph between GDP per capita with dropout rates in education (primary, upper primary and secondary) for the year 2014-2015 for each state. 

```python
# Since There is huge difference in values of GDP per capita and the % Drop out rates, so calculating the Log base 10 value of GDP per capita
Final_DF['GDP_per_capita_log']=np.log2(Final_DF['Per_Capita_GSDP'])
Final_DF=Final_DF.fillna(0)
Final_DF.sort_values(by='Per_Capita_GSDP',ascending=False)
```

```python
DF_GDP_PC_DRP_RT=Final_DF.loc[:,['GSDP','Primary','Secondary','Tertiary','Primary_-_2014-2015','Upper_Primary_-_2014-2015','Secondary_-_2014-2015']]
#Final_DF=temp.loc[:,['Primary','Secondary','Tertiary','GSDP','Per_Capita_GSDP','Population','Primary_-_2014-2015','Upper_Primary_-_2014-2015','Secondary_-_2014-2015','Senior_Secondary_-_2014-2015']]
```

```python
#print(type(list(Final_DF['GDP_per_capita_log'])))
#plot_graph(list(Final_DF['GDP_per_capita_log']),list(Final_DF['Primary_-_2014-2015']))
#plot_graph(list(Final_DF['GDP_per_capita_log']),list(Final_DF['Upper_Primary_-_2014-2015']))
#plot_graph(list(Final_DF['GDP_per_capita_log']),list(Final_DF['Secondary_-_2014-2015']))
Final_DF=Final_DF.sort_values('Per_Capita_GSDP',ascending=True)
import matplotlib.pyplot as plt
x_GDP_per_capita=list(Final_DF['GDP_per_capita_log'])
y_prim_drp= list(Final_DF['Primary_-_2014-2015'])
y_up_prim_drp=list(Final_DF['Upper_Primary_-_2014-2015'])
y_sec_drp=list(Final_DF['Secondary_-_2014-2015'])

m1,c1=np.polyfit(x_GDP_per_capita,y_prim_drp,1)
m2,c2=np.polyfit(x_GDP_per_capita,y_up_prim_drp,1)
m3,c3=np.polyfit(x_GDP_per_capita,y_sec_drp,1)
plt.plot(x_GDP_per_capita,m1*np.array(x_GDP_per_capita)+c1,'r',label="PRIMARY DROP RATE")
plt.plot(x_GDP_per_capita,m2*np.array(x_GDP_per_capita)+c2,'g',label="UPPER PRIMARY DROP RATE")
plt.plot(x_GDP_per_capita,m3*np.array(x_GDP_per_capita)+c3,'b',label="SECONDRY DROP RATE")
plt.title("Correlation between per-capita GDP and Drop out rate at Various level of education")
plt.legend()
#plt.set_xticklabels(labels)
plt.show()
```
